# 2.输入一个时间值s，它是距当日午夜的秒值，计算目前的时间，时间按 HH:MM:SS 格式 输出。



```python
def convert(s):
    left = 3600*24 - s
    s = left%60
    m = left//60
    h = 0
    if m > 59 :
        m = m%60
        h = m//60
    if h < 10:
        h = '0'+ str(h)
    if m < 10:
        m = '0' + str(m)
    if s < 10:
        s = '0' + str(s)
    return str(h)+":"+str(m)+":"+str(s)
        

#     return "23:00:00"

print(convert(67889))
print(convert(65467))
print(convert(86578))
print(convert(75675))
print(convert(86578))
```
