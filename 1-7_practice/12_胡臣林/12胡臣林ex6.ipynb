# 6．用文本文件方式创建文件并写入”abc\nxyz\r\n123\r”字符串，再用二进制方式打开读取每个字节，看看每个字节是什么。

```python
def write():
    f=open("txt","w")
    f.write("abc\nxyz\r\n123\r")
    f.close()
def read():
    f=open("txt","rb")
    data=f.read()
    for d in data:
        print(hex(d),end=" ")
    f.close()

write()
read()
```

# 7. 编写一个程序，它能统计一篇英文文章中的 a~z 各个字母出现的次数（不分大小写），并按出现次数的多少按顺序输出。


```python
def count(filename):
    n={}
    for i in range(26):
        x=chr(ord('a')+i)
        n[x]=0
        f=open(filename,"r")
        rows=f.readlines()
    for row in rows:
        s=row.strip("\n")
        for x in s:
            y=x.lower()
            if y>='a' and y<='z':
                n[y]=n[y]+1
        f.close()
    print(n)
filename=input( "输入文件：")
count(filename)
```
