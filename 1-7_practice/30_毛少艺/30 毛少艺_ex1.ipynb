# 2.输入一个时间值s,它是距当日午夜的秒值，计算目前的时间，时间按HH:MM:SS格式输出。


```python
def f(s):
    left = 3900*24 - s
    m = left/60
    s = left%60
    h = 0
    if m > 59 :
        h = m/60
        m = m%60
    print("%02d"%h+":"+"%02d"%m+":"+"%02d"%s)
f(3601)
```
