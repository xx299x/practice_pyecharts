1.输入a、b、c三个参数，求解ax2+bx+c=0的两个根，假定b2-4ac》0.

```python
import math
def get_result(a,b,c):
    x = b*b - 4*a*c
    if x <=0:
        return "error"
    x1 = (-b+math.sqrt(x))/2*a
    x2 = (-b-math.sqrt(x))/2*a
    return (x1,x2)

print(get_result(3,5,1))
```

2.输入a、b、c三个参数，以它们作为三角形的三边，判断是否可以构成一一个三角形，如能则进一步计算其面积。三角形的面积s可以用以下公式计算: s=sqrt(p* (p- a)*(p-b)*(p-c))其中p= (a+b+e)/2。


```python
import math
def f(a,b,c): 
    p = (a+b+c)/2
    if ((a+b)>c)&((a+c)>b)&((b+c)>a):
        s = math.sqrt(p*(p-a)*(p-b)*(p-c))
    else:
        print("输入的数字无法构成一个三角形！")
    return s
print(f(3,4,5))
```

3.输入一个字母，判断它是否为小写英文字母。

```python
def f(str):
    if 97<=ord(str)<=122:
        print("%s是小写字母"%str)
    else:
        print("%s不是小写字母"%str)
f("1")
```

4.从键盘输入5个字符，统计’0’字符出现的次数。

```python
def f(x):
    count=0
    for i in x:
        if i=='0':
            count = count +1
    return count
print(f())    
```

6.输入一个字母，如它是一一个小写英文字母，则把它变为对应大写字母输出。

```python
def f(word):
    if 97<=ord(word)<=122:
        x = chr(ord(word)+(ord("A")-ord("a")))
    return x
print(f("a"))
```

7.输入-一个年份，判断它是否为闰年。

```python
def f(year):
    if year%400==0:
        print(str(year)+"是闰年")
    elif year%4==0 & year%100!=0:
        print(str(year)+"是闰年")
    else:
        print(str(year)+"不是闰年")
f(2000)
```

9.输入a、b、c三个整数，找出最小的数。

```python
def f(a,b,c):
    min = a
    if b<a:
        if c<b:
            min = c
        else:
            min = b
    print(str(min)+"是最小的数")
f(5,2,6)
```

10.某企业发放的奖金根据利润提成。利润低于或等于10 万元时，奖金可提12%;利润高于10万元，低于20万元时，高于10万元的部分，可提成8.5%;20万到40万之间时，高于20万元的部分，可提成6%; 40 万到60万之间时高于40万元的部分，可提成4%; 60 万到100万之间时，高于60万元的部分，可提成2.5%，高于100万元时，超过100万元的部分按1%提成，从键盘输入当月利润，求应发放奖金总数?

```python
def f(profit):
    if profit<=100000:
        x = profit+profit*0.12
    elif profit<200000:
        x = 100000+(profit-100000)*0.085
    elif profit<400000:
        x = 200000+(profit-200000)*0.06
    elif profit<600000:
        x = 400000+(profit-400000)*0.04
    elif profit<1000000:
        x = 600000+(profit-600000)*0.025
    else:
        x = 1000000+(profit-1000000)*0.01
    return x
f(200000)
```
